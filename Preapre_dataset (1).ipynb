```python
import xml.etree.ElementTree as ET

tree = ET.parse(words_xml_path)
root = tree.getroot()

for image in root:
    img_name = image.find('imageName').text

    for bbs in image.findall('taggedRectangles'):
        for bb in bbs.findall('taggedRectangle'):
            bbox = [
                int(float(bb.attrib['x'])),
                int(float(bb.attrib['y'])),
                int(float(bb.attrib['width'])),
                int(float(bb.attrib['height']))
            ]

            text = bb.find('tag').text
            print(img_name, text, bbox)

    break
```

```python
def extract_data_from_xml(path):
    tree = ET.parse(path)
    root = tree.getroot()

    # Danh sách lưu dữ liệu cho TOÀN BỘ dataset
    image_paths = []               # lưu tên
    image_sizes = []               # lưu kích thước
    image_labels = []              # lưu nhãn
    bounding_boxes = []            # lưu bounding box

    for image in root:
        bbs_of_image = []
        labels_of_image = []

        for bbs in image.findall("taggedRectangles"):
            for bb in bbs:
                # check non-alphabet and non-number
                if not bb[0].text.isalnum():
                    continue

                # Bỏ qua các ký tự có dấu (ngoài bảng ASCII)
                if "é" in bb[0].text.lower() or "ñ" in bb[0].text.lower():
                    continue

                bbs_of_image.append(
                    [
                        float(bb.attrib["x"]),
                        float(bb.attrib["y"]),
                        float(bb.attrib["width"]),
                        float(bb.attrib["height"]),
                    ]
                )
                labels_of_image.append(bb[0].text.lower())

        # Store
        image_paths.append(image[0].text)
        image_sizes.append((int(image[1].attrib["x"]), int(image[1].attrib["y"])))
        bounding_boxes.append(bbs_of_image)
        image_labels.append(labels_of_image)

    return image_paths, image_sizes, image_labels, bounding_boxes

image_paths, image_sizes, image_labels, bounding_boxes = extract_data_from_xml(words_xml_path)
```

```python
for path, size, label, bbox in zip(image_paths, image_sizes, image_labels, bounding_boxes):
    print(path, size, label, bbox)
    break
```
